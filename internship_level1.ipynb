Explore the distribution of categorical variables like "Country Code","City" and "Cuisines".

```python
#EXPLORE THE DITRIBUTION OF COUNTRY CODE
plt.figure(figsize=(7,5))
sns.countplot(x='Country Code',data=data1)
plt.title("Distribution of restaurants according to country code")
plt.xlabel('country code')
plt.ylabel('No. of restaurants')
plt.show()
```

Highest number of restaurants are there in country code 1 and the second highest in country code 216


```python
#EXPLORE THE DITRIBUTION OF CITIIES
plt.figure(figsize=(10,7))
sns.countplot(x='City',data=data1,order=data1['City'].value_counts().head(40).index)
plt.xticks(rotation=75)
plt.title("Distribution of restaurants according to cities")
plt.xlabel('city')
plt.ylabel('No. of restaurants')
plt.show()
```

Highest number of restaurants are there in New Delhi and the second highest in Gurgoan

```python
#EXPLORE THE DITRIBUTION OF cuisines
plt.figure(figsize=(15,5))
sns.countplot(x='Cuisines',data=data1,order=data1['Cuisines'].value_counts().head(40).index)
plt.xticks(rotation=90)
plt.title("Distribution of restaurants according to cuisines")
plt.xlabel('cuisine')
plt.ylabel('No. of restaurants')
plt.show()
```

Top cuisine in many restaurants is north indian and chinese

```python
#IMPORT LIBRARIES
import geopandas as gpd
from geopandas import GeoDataFrame
from shapely.geometry import Point
```

Visualize the locations of restaurants on a map using latitude and longitude information

```python
#VISUASLIZING THE LOCATIONS OF RESTAURANTS ON A MAP USING LATITUDE AND LONGITUDE INFORMATION.
gll=gpd.GeoDataFrame(data1,geometry=gpd.points_from_xy(data1.Longitude,data1.Latitude))
map=gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gll.plot(ax=map.plot("continent",figsize=(20,20)),marker='*',color='red')
plt.show()
```

Analyze the distribution of restaurants across different cities or countries.


```python
#EXPLORE THE DITRIBUTION OF RESTAURANTS ACROSS DIFFERENT CITIIES.
plt.figure(figsize=(10,7))
sns.countplot(x='City',data=data1,order=data1['City'].value_counts().head(25).index)
plt.xticks(rotation=75)
plt.title("Distribution of restaurants according to cities")
plt.xlabel('city')
plt.ylabel('No. of restaurants')
plt.show()
```

Determine if there is any correlation between the restaurant's location and its rating.

```python
#DETERMINING IF THERE IS CORRELATION BETWEEN THE RESTAURANTS LOCATION AND ITS RATING.
cmtrix=data1[['Latitude','Longitude','Aggregate rating']].corr()
cmtrix
```
